```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
urunler=pd.read_excel("urunler.xlsx")
urunler=pd.DataFrame(urunler)
```

```python
last=urunler.iloc[0:13]
last
```

```python
last.rename(columns={'Unnamed: 0':"quarter"},inplace=True)
last.rename(columns={'Motor vehicle and parts':"Motor vehicle and parts"},inplace=True)
last.rename(columns={'Furniture and\nhome\nfurnishings':"Furniture and home furnishings"},inplace=True)
last.rename(columns={'Building materials,\ngarden equipment\nand supplies':"Building materials, garden equipment and supplies"},inplace=True)
last.rename(columns={'Clothing and\nclothing\naccessories':"Clothing and clothing accessories"},inplace=True)
last.rename(columns={'Food and\nbeverage':"Food and beverege"},inplace=True)
last.rename(columns={'Health and\npersonal\ncare':"Health and personal care"},inplace=True)
last.rename(columns={'Sporting goods, hobby\nmusical instrument,\nand books':"Sporting goods, hobby musical instrument and books"},inplace=True)
```
